```python
 # Convert to datetime 
data = {
    'name': [
        "Alice", "Bob", "Charlie", "David", "Eva",
        "Frank", "Grace", "Helen", "Ian", "Jane"
    ],
    "salary": [
        35000, 42000, 39000, 41000, 37000,
        43000, 40000, 39500, 38000, 40500
    ],
    "date": [
        "2016-06-11 08:15:00",
        "2017-06-11 08:45:00",
        "2018-06-11 09:00:00",
        "2019-06-11 08:55:00",
        "2020-06-11 09:10:00",
        "2021-06-11 08:30:00",
        "2022-06-11 09:05:00",
        "2023-06-11 08:40:00",
        "2024-06-11 08:50:00",
        "2025-06-11 09:15:00"
    ]

}
df = pd.DataFrame(data)
df['date'] = pd.to_datetime(df['date']) 
df.set_index('date', inplace=True) # Trend analysis 
df['salary'].plot(figsize=(12, 6)) 
df['salary'].rolling(window=30).mean().plot() # Seasonality 
df.groupby(df.index.month)['salary'].mean().plot(kind='bar')
df
```

```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 

# Create sample data 
data = { 'sales': [100, 150, 200, 120, 180], 'marketing': [20, 30, 40, 25, 35], 'satisfaction': [8.5, 9.2, 9.8, 8.8, 9.5], 'price': [50, 45, 40, 48, 42] } 
df = pd.DataFrame(data) # Calculate correlation matrix

correlation_matrix = df.corr() 
print(correlation_matrix) # Visualize with heatmap 

plt.figure(figsize=(8, 4)) 

sns.heatmap(correlation_matrix) 
sns.heatmap(correlation_matrix, annot=True)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm',  center=2)

plt.title('Correlation Matrix Heatmap') 
plt.show()
```
