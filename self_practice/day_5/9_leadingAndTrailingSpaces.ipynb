```python
# You have a DataFrame where some string columns have leading/trailing whitespaces. How would you clean all string columns at once?

import pandas as pd 
data = {
    'name': [' Alice ', 'Bob', '  Charlie', 'David  '],
    'city': [' New York', 'London ', '  Paris ', 'Berlin'],
    'score': [85, 90, 78, 92]  # numeric column for contrast
}

df = pd.DataFrame(data) 

# df['city'] = df['city'].str.strip() 

df 

df[df.select_dtypes(include='object').columns] = df.select_dtypes(include='object').apply(lambda x: x.str.strip())

df 
```

```python
df = pd.DataFrame(data) 

df[df.select_dtypes('object').columns] = df.select_dtypes('object').apply(lambda x : x.str.strip())
df
```
