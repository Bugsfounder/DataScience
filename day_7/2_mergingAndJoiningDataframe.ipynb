```python
import pandas as pd
customers = pd.DataFrame({'customer_id':[1,2,3,4],'name':["Alice", "Bob","Charlie", 'Diana'],"city": ["NYC", "UA", "Chicago", "India"],
    "salary": [50000, 60000, 70000, 80000]}) 

orders = pd.DataFrame( {'order_id' : [101,102,103,104,105], 'customer_id':[1,2,1,3,5], "product":['laptop','phone','tablet','watch','camera'], 'amount':[1200,800,550,300,900]})
```

```python
# inner join --> intersection
result = pd.merge(customers,orders, on='customer_id', how='inner')

print(result)
```

```python
# outer join
result = pd.merge(customers,orders, on='customer_id',how="outer")

print(result)
```

```python
# left join
result = pd.merge(customers,orders, on='customer_id',how="left")

print(result)
```

```python
# right join
result = pd.merge(customers,orders, on='customer_id',how="right")

print(result)
```

```python
# join method --> join()
# customers.set_index('customer_id')
# orders.set_index('customer_id')
result = customers.set_index('customer_id').join(orders.set_index('customer_id'), how="outer")

result
```

```python
result1 = customers.set_index('customer_id').join(orders.set_index('customer_id'), how="right")
result2 = customers.set_index('customer_id').join(orders.set_index('customer_id'), how="left")
result3 = customers.set_index('customer_id').join(orders.set_index('customer_id'), how="inner")
print(result1)
print(result2)
print(result3)
```

```python
all_customers = pd.concat([customers,orders], ignore_index=True)
all_customers
```

```python
all_customers = pd.concat([customers,orders], ignore_index=True, axis=1)
all_customers
```
