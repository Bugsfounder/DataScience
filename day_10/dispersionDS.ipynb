```python
# - How does the variance of exam scores differ between two university courses? Compare the score distributions of two subjects to analyze which one has more variation.
course_A = [70, 75, 80, 85, 90]
course_B = [60, 65, 90, 95, 100]

import numpy as np
var_A = np.var(course_A, ddof=1)
var_B = np.var(course_B, ddof=1)


var_A

var_B
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

prices = np.random.normal(loc=100, scale=10, size=100)  # Normal-like
# prices = np.append(prices, [200, 220])  # To add skew/outliers

sns.histplot(prices, kde=True)
plt.title("Stock Prices Distribution")
plt.show()
```
