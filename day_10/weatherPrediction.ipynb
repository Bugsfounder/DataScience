```python
data = {
    'datetime': ['2025-06-01 00:00', '2025-06-01 01:00', '2025-06-01 02:00', '2025-06-01 03:00', '2025-06-01 04:00'],
    'temperature_C': [24.5, 24.0, 23.7, None, 23.2],
    'humidity_%': [78, 80, None, 85, 87],
    'pressure_hPa': [1012, 1011, 1010, 1009, None],
    'wind_speed_kmph': [12.0, 11.5, None, 10.0, 9.8],
    'rainfall_mm': [0.0, 0.0, 0.2, 0.5, None]
}

# datetime: Timestamp of hourly data
# temperature_C: Temperature in Celsius (has 1 missing)
# humidity_%: Humidity percentage (1 missing)
# pressure_hPa: Atmospheric pressure (1 missing)
# wind_speed_kmph: Wind speed (1 missing)
# rainfall_mm: Rainfall amount (1 missing)


# âœ… Use Cases:
    # Time series prediction (next hour temperature, rainfall, etc.)
    
    # Handle missing values (interpolation)
    
    # Visualizations (temp/humidity trends)

```

```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 

# handeling missing values 

df = pd.DataFrame(data) 
df['temperature_C'] = df['temperature_C'].interpolate()
df['humidity_%'] = df['humidity_%'].interpolate() 
df['pressure_hPa'] = df['pressure_hPa'].interpolate() 
df['wind_speed_kmph'] = df['wind_speed_kmph'].interpolate()
df['rainfall_mm'] = df['rainfall_mm'].interpolate()

df['datetime'] = pd.to_datetime(df['datetime'])

plt.figure(figsize=(10,8))
# plt.scatter(df['datetime'], df['temperature_C'], label="h1")
# plt.plot(df['datetime'],df['humidity_%'], label="h2" )
plt.plot(df['temperature_C'], df['humidity_%'], label="h3")
plt.scatter(df['temperature_C'], df['humidity_%'])
plt.stem(df['temperature_C'], df['humidity_%'])
plt.title("temp/humidy visualizer")

plt.legend()
```
