```python
# https://raw.githubusercontent.com/thepycoach/data_preprocessing/main/datasets/StudentsPerformance.csv
# https://raw.githubusercontent.com/thepycoach/data_preprocessing/main/netflix_titles.csv

import pandas as pd
import numpy as np

# read csv
url = "https://raw.githubusercontent.com/thepycoach/data_preprocessing/main/netflix_titles.csv"
url = "https://raw.githubusercontent.com/thepycoach/data_preprocessing/main/datasets/StudentsPerformance.csv"
data = pd.read_csv(url)
df = pd.DataFrame(data)
```

```python
# dropna()
# dropping null values
data_dropped_na = df.dropna()

'''
scale 
ffill
bfill
mean, median, mode
dictionary --> different values for different column
'''

# fillna()
# filling null values 
fill_scale = df.fillna(10)
fill_forward = df.ffill()
fill_backward = df.bfill()

# fill mean meadian mode
mean_fill  = df["math score"].fillna(df["math score"].mean())
meadian_fill = df['math score'].fillna(df['math score'].median())
mode_fill = df['math score'].fillna(df['math score'].mode())
```

```python
# df.replace({oldv: newv})
# Real-world Use Case:
# Cleaning messy data like converting all 'N/A', 'missing', or 'null' to np.nan.

# we can use replace function to replace a null value which is not by default treated as null, like false or may be 0
df.replace({'False' : np.nan})

# replaec a value a value in data with different value for all occurenc
data_rep_72_0 = df.replace({72:0})
data_rep_in_math_core_col = df["math score"].replace({72:0})
data_rep_in_math_core_col
```

```python
# ðŸ”¹ Why inplace=True?

# inplace=True modifies the original DataFrame directly, without needing reassignment.
df.replace({72:0}, inplace=True) # this will directly modify the dataframe
df
```
