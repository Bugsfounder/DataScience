```python
import pandas as pd

df = pd.DataFrame({
    'Department': ['HR', 'IT', 'HR', 'IT', 'HR'],
    'Salary': [40000, 50000, 42000, 52000, 41000]
})
# Q: How can you find the average salary for each department using Groupby aggregation?

df.groupby("Department")['Salary'].mean()
```
