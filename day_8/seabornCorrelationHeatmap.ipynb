```python
np.random.seed(42)
n_samples = 500

data = {
    'price': np.random.lognormal(12, 0.4, n_samples).round(2),
    'sqft': np.random.normal(1800, 600, n_samples).round(),
    'bedrooms': np.random.choice([1, 2, 3, 4, 5], n_samples, p=[0.1, 0.2, 0.4, 0.2, 0.1]),
    'bathrooms': np.random.choice([1, 1.5, 2, 2.5, 3], n_samples),
    'age': np.random.randint(1, 50, n_samples),
    'distance_to_city': np.random.exponential(10, n_samples).round(1),
    'has_garage': np.random.binomial(1, 0.7, n_samples),
    'school_rating': np.random.choice(range(1, 11), n_samples)
}

df = pd.DataFrame(data)

corr = df.select_dtypes('number').corr()

corr

sns.heatmap(corr)
```
