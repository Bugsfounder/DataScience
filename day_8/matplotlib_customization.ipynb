```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.figure(figsize=(10, 6))
plt.plot(x,y)


plt.xlabel('x-label')
plt.ylabel('y-label')
plt.ylabel('y-label', fontsize=12, fontweight='bold')

print(x)
```

```python
x  = np.linspace(1,9,1000)
y = np.sin(x)

plt.plot(x,y, label="red")
plt.scatter(x,y, color='pink')

plt.legend(loc='upper right')
```

```python
# plt.xlabel('label', fontsize=14, fontweight='bold', color='pink', fontfamily="ariel")
plt.xlabel('label', fontsize=14, fontweight='bold', color='pink')
plt.title("Title of the graph")
plt.suptitle("Sup Title")
```

```python
plt.figure(figsize=(10,6))
plt.plot(x,y, label="six(x)", linewidth=2)
plt.plot(x,np.cos(x), label='cos(x)', linewidth=2)
plt.plot(x,np.tan(x), label='tan(x)', linewidth=2)
plt.legend()
plt.legend(loc='upper right')
plt.legend(bbox_to_anchor=(1,1), loc='upper right')
plt.legend(bbox_to_anchor=(1.2,1), loc='upper right')

plt.text(5,0.5, 'peek value', fontsize=12,color='red')
plt.text(3,-0.4, 'Minimum point', fontsize=12, bbox=dict(boxstyle='round', pad=0.3, facecolor='yellow'))
```
